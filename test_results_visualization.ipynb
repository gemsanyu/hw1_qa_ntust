```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("test_results.csv")
```

```python
# Optional: format model names nicely
df['model'] = df['model'].str.replace('_', ' ').str.title()

# Set up a consistent style
metrics = ['rmse', 'mse', 'mae']

# Plot each metric
for metric in metrics:
    plt.figure(figsize=(10, 5))
    df_sorted = df.sort_values(by=metric)
    plt.bar(df_sorted['model'], df_sorted[metric], color='blue')
    plt.title(f'{metric.upper()} by Model')
    plt.ylabel(metric.upper())
    plt.yscale("log")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()# Optional: format model names nicely
df['model'] = df['model'].str.replace('_', ' ').str.title()
```
